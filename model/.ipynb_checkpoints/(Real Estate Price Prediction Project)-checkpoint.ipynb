```python
#take the average of the range and input it
def convert_sqft_to_num(x):
    token=x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
         return float(x)
    except:
        return None
    
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_unit)
        st = np.std(subdf.price_per_unit)
        reduced_df = subdf[(subdf.price_per_unit>(m-st)) & (subdf.price_per_unit<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_pps_outliers(df6)
df7.shape
```

```python
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.BHK==2)]
    bhk3 = df[(df.location==location) & (df.BHK==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()
    
plot_scatter_chart(df7,"Rajaji Nagar")
```

```python
#chatgpt function instead of creating ur own use it
import matplotlib.pyplot as plt

def plot_location_scatter(df, location, x_col='total_sqft', y_col='price', color='blue'):
    """
    Plots a scatter plot of a specific location from the DataFrame.

    Parameters:
        df (pd.DataFrame): Your dataset
        location (str): The location to filter by
        x_col (str): Column for the x-axis (default: 'total_sqft')
        y_col (str): Column for the y-axis (default: 'price')
        color (str): Point color
    """
    loc_df = df[df['location'] == location]

    if loc_df.empty:
        print(f"No data found for location: {location}")
        return

    plt.figure(figsize=(20, 10))
    plt.scatter(loc_df[x_col], loc_df[y_col], color=color, s=60, alpha=0.6, edgecolors='k')
    plt.title(f"{y_col} vs {x_col} for {location}", fontsize=16)
    plt.xlabel(x_col, fontsize=14)
    plt.ylabel(y_col, fontsize=14)
    plt.grid(True)
    plt.show()
plot_scatter_chart(df7,"Hebbal")
```

```python
#We should also remove properties where for same location, the price of (for example) 3 bedroom apartment is less than 2 bedroom apartment (with same square ft area). What we will do is for a given location, we will build a dictionary of stats per bhk, i.e.

#{
 #   '1' : {
    # 'mean': 4000,
      #  'std: 2000,
       # 'count': 34
    #},
    #'2' : {
     #   'mean': 4300,
      #  'std: 2300,
       # 'count': 22
    #},    
#}
#Now we can remove those 2 BHK apartments whose price_per_sqft is less than mean price_per_sqft of 1 BHK apartment
```

```python
import numpy as np

def remove_BHK_outliers(df):
    exclude_indices = np.array([])

    for location, location_df in df.groupby('location'):
        bhk_stats = {}

        # Calculate mean, std, and count for each BHK level
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_unit),
                'std': np.std(bhk_df.price_per_unit),
                'count': bhk_df.shape[0]
            }

        # Compare current BHK prices with previous BHK level (BHK-1)
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                condition = bhk_df.price_per_unit < stats['mean']
                exclude_indices = np.append(exclude_indices, bhk_df[condition].index.values)

    return df.drop(exclude_indices, axis='index')
df8= remove_BHK_outliers(df7)
df8.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import numpy as np

def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False],
                'positive': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],  # updated from 'mse'
                'splitter': ['best', 'random']
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr.predict([x])[0]
```
